```python
import numpy as np
from scipy.stats import t, ttest_ind
import matplotlib.pyplot as plt

# генерируем данные для двух выборок
np.random.seed(42)
sample1 = (18.15, 18.32, 18.28, 18.19, 18.65, 18.33, 18.44, 18.79, 19.18, 18.98, 18.66, 18.99, 18.79, 18.96, 18.30)
sample2 = np.random.uniform(18.3, 18.9, size=15)

# проводим односторонний t-тест
t_stat, p_value = ttest_ind(sample1, sample2)
print("t-статистика =", t_stat)
print("p-value =", p_value)

# строим распределение t-статистики
df = len(sample1) + len(sample2) - 2
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 100)
y = t.pdf(x, df)
plt.plot(x, y)

# отмечаем критическое значение t-статистики на уровне значимости 0.05
t_crit = t.ppf(0.95, df)
plt.axvline(x=t_crit, color='r', linestyle='--')

# отмечаем наблюдаемую t-статистику
plt.axvline(x=t_stat, color='g', linestyle='--')

plt.xlabel('t')
plt.ylabel('pdf(t)')
plt.show()
```

```python
import random
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# генерируем данные для двух выборок
sample1 = [random.uniform(18.3, 18.9) for _ in range(15)]
sample2 = [random.uniform(18.3, 18.9) for _ in range(15)]

# строим график распределения выборок
plt.hist(sample1, alpha=0.5, label='Sample 1')
plt.hist(sample2, alpha=0.5, label='Sample 2')
plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(sample2), color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()

# проводим односторонний t-тест
t_stat, p_value = stats.ttest_ind(sample1, sample2)
print('t-статистика:', t_stat)
print('p-значение:', p_value)

# строим распределение t-статистики с отмеченным критическим значением
df = len(sample1) + len(sample2) - 2
alpha = 0.05
t_critical = stats.t.ppf(alpha, df)
x = np.linspace(-5, 5, 1000)
y = stats.t.pdf(x, df)
plt.plot(x, y)
plt.axvline(t_critical, color='red', linestyle='dashed', linewidth=2)
plt.axvline(t_stat, color='green', linestyle='dashed', linewidth=2)
plt.fill_between(x, 0, y, where=x<=t_critical, color='gray', alpha=0.5)
plt.title('Распределение t-статистики')
plt.show()

# строим диаграмму ящик с усами
plt.boxplot([sample1, sample2])
plt.xticks([1, 2], ['Sample 1', 'Sample 2'])
plt.title('Диаграмма ящик с усами')
plt.show()
```

```python
import random
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# генерируем данные для двух выборок
sample1 = [random.uniform(18.3, 18.9) for _ in range(15)]
sample2 = [random.uniform(18.3, 18.9) for _ in range(15)]

# строим график распределения выборок
plt.hist(sample1, alpha=0.5, label='Sample 1')
plt.hist(sample2, alpha=0.5, label='Sample 2')
plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(sample2), color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()

# проводим правосторонний t-тест
t_stat, p_value = stats.ttest_ind(sample1, sample2, alternative='greater')
print('Правосторонний t-тест:')
print('t-статистика:', t_stat)
print('p-значение:', p_value/2) # делим p-значение на 2, так как это правосторонний тест

# строим распределение t-статистики с отмеченным критическим значением
df = len(sample1) + len(sample2) - 2
alpha = 0.05
t_critical = stats.t.ppf(1-alpha, df)
x = np.linspace(-5, 5, 1000)
y = stats.t.pdf(x, df)
plt.plot(x, y)
plt.axvline(t_critical, color='red', linestyle='dashed', linewidth=2)
plt.axvline(t_stat, color='green', linestyle='dashed', linewidth=2)
plt.fill_between(x, 0, y, where=x>=t_critical, color='gray', alpha=0.5)
plt.title('Распределение t-статистики')
plt.show()

# строим диаграмму ящик с усами
plt.boxplot([sample1, sample2])
plt.xticks([1, 2], ['Sample 1', 'Sample 2'])
plt.title('Диаграмма ящик с усами')
plt.show()
```
